```python
# Test R environment with rpy2
import rpy2.robjects as ro
from rpy2.robjects.packages import importr
from rpy2.robjects.vectors import FloatVector
import matplotlib.pyplot as plt
from IPython.display import display, Image

# Activate pandas2ri for automatic conversion
from rpy2.robjects import pandas2ri
pandas2ri.activate()

# Check if ggplot2 is installed, and install it if not
ro.r('''
if (!requireNamespace("ggplot2", quietly = TRUE)) {
    install.packages("ggplot2", repos = "http://cran.us.r-project.org")
}
''')

# Import ggplot2
ggplot2 = importr('ggplot2')

# Run a simple calculation in R
result = ro.r('2 + 2')
print(f"Simple calculation in R (2 + 2): {result[0]}")

# Create a simple plot in R
ro.r('''
library(ggplot2)
data <- data.frame(x = c(1, 2, 3, 4, 5), y = c(1, 4, 9, 16, 25))
p <- ggplot(data, aes(x = x, y = y)) +
    geom_point() +
    geom_line() +
    ggtitle("Test Plot: y = x^2")
ggsave("test_plot.png", plot = p, width = 5, height = 4)
''')

# Display the plot in Jupyter Notebook
display(Image(filename="test_plot.png"))
```
